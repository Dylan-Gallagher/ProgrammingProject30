```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_file.csv' with the actual name of your CSV file
data = pd.read_csv('flights2k_new.csv')

# Count the flights from each origin
origin_counts = data['ORIGIN'].value_counts()

# Create the bar chart
plt.figure(figsize=(14, 6))
origin_counts.plot(kind='bar')
plt.xlabel('Airport Codes')
plt.ylabel('Number of Flights')
plt.title('Distribution of Flights by Origin Airport')

# Display the bar chart
plt.show()
```
